```python
import pandas as pd
import seaborn as sns
```

```python
penguins = sns.load_dataset("penguins")
penguins.head()
```

```python
# Import house price data
houses = pd.read_csv('house_train.csv')
houses['date_sold'] = houses.apply(lambda row: str(row.YrSold) + '/' + str(row.MoSold), axis='columns')
houses.date_sold = pd.to_datetime(houses.date_sold)
houses.head()
```

```python
# Your code here
sns.relplot(kind='line', data=houses, x='date_sold', y='SalePrice')
```
